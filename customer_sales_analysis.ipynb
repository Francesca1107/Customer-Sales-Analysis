Customer & Sales Analysis

## Objective
The goal of this project is to analyze customer behavior and identify key patterns in sales and order frequency.

## Dataset
The dataset contains information about orders, including custerm ID, order ID, city, product category, and sales.

```python
import pandas as pd
train = pd.read_csv('/Users/francescalovecchio/Downloads/train.csv')
```

```python
train.head()
```

```python
#calculate total sales for each customer
tot_sale = train.groupby('Customer ID')['Sales'].sum().sort_values(ascending = False)
```

```python
#count number of unique orderds per customer
tot_ord = train.groupby('Customer ID')['Order ID'].nunique().sort_values(ascending=False)
```

```python
#combine sales and orders data into a single dataframe
ord_sale = pd.concat([tot_sale, tot_ord], axis = 1).sort_values(ascending = False, by = 'Sales')
```

```python
#identify customers with high number of orders
ord_sale.sort_values(ascending = False, by = 'Order ID')
```

## Customer Behavior Insights

- Some customers generate high revenue with a small number of large orders
- Other customers place frequent orders but with lower average value
- This suggests different customer segments that may require different strategies

```python
import matplotlib.pyplot as plt
plt.scatter(x = ord_sale['Order ID'], y = ord_sale['Sales'])
plt.xlabel('Orders')
plt.ylabel('Sales')
plt.title('Behavior of Customer')
plt.grid()
plt.show()
```

## Data Visualization Insights

- A small number of customers (outliers) generate significantly higher sales or orders compared to the rest
- Most customers are clustered in the lower range of sales and orders
- This indicates that revenue is concentrated among a limited number of high-value customers

## Key Insights

- A small number of customers generate most of the revenue
- Some customers place many orders but spend less per order
- Customer behavior varies significantly across the dataset

## Business Recommendations

- Focus on high-value customers to maximize revenue
- Develop strategies for frequent customers to increase their average order value
- Identify opportunities to convert low-value customers into higher-value segments
